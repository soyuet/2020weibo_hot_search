## 3. 情感分析

#### 微博热搜是消极还是积极的内容？ 
看看2020年的热搜包含的情感。这里要引入一个第三库（snownlp），这个库可以直接对文本进行情感分析。返回的是积极性的概率，数值介于[0,1]之间。

```python
from snownlp import SnowNLP
```

```python
# 把所有内容拼成一个字符串
for word in clean_cut_word:
    text1 = ','.join(word) 
s1 = SnowNLP(text1)
print(s1.sentiments)
```

#### 得到的值为0.316，可以看出2020年的微博热搜的用语，整体上是偏向于消极的。
#### 那对于具体每一条热搜来说呢？

```python
sentiments_each_weibo =[]
for word in title_list:
    temp = SnowNLP(word)
    sentiments_each_weibo.append(temp.sentiments)
sentiments_each_weibo
```

```python
positive_content=[]
negative_content=[]
for y in sentiments_each_weibo:
    if y >= 0.5:
        positive_content.append(y)
    else:
        negative_content.append(y)
```

```python
print('积极的微博数量为：'+str(len(positive_content)))
print('消极的微博数量为：'+str(len(negative_content)))
```

#### 具体看每一条微博内容，可以看出其中含积极情感的微博数量是2倍于消极的。看来确实是正面的事情更适合热搜。

```python
# df3 存储情感指数、热搜话题、和时间

# df3 = pd.DataFrame(title_list,columns=['title'])
df3 = pd.DataFrame(df1['date'].tolist(),columns=['date'])
df3 = pd.concat([df3, pd.DataFrame(sentiments_each_weibo,columns=['sentiments'])],axis=1)
df3 
```

#### 接下来将每一天微博热搜的情感指数和搜索量分别进行加和，再将情感的数据，与搜索量的数据，放在一个图表中，看看全年微博热搜的情感是如何变化的。
